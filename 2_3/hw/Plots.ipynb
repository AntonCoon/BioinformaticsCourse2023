```python
import networkx as nx
import matplotlib.pyplot as plt
import collections
import ast
import networkx as nx
from collections import defaultdict

def make_a_plot(deg, cnt):
  fig, ax = plt.subplots()
  plt.bar(deg, cnt, width=0.80, color='b')
  plt.ylabel("Количество")
  plt.xlabel("Степень")
  ax.set_xticks([d + 0.4 for d in deg])
  ax.set_xticklabels(deg)
  ax.set_xticks([d + 0.4 for d in deg])
  ax.set_xticklabels(deg)
  plt.show()



with open('/content/drive/MyDrive/магистратура/intermediate_calculations.txt', 'r') as f:
    for line in f:
        if line.startswith('graph:'):
            graph_str_1 = line.split('graph: ')[1]
            graph_1 = eval(graph_str_1)
            G_1 = nx.Graph(graph_1)
            degree_sequence_1 = sorted([d for n, d in G_1.degree()], reverse=True)
            degreeCount_1 = collections.Counter(degree_sequence_1)
            deg_1, cnt_1 = zip(*degreeCount_1.items())
            continue
        if line.startswith('compressed_graph:'):
            graph_str_2 = line.split('compressed_graph: ')[1]
            graph_2 = eval(graph_str_2)
            G_2 = nx.Graph(graph_2)
            degree_sequence_2 = sorted([d for n, d in G_2.degree()], reverse=True)
            degreeCount_2 = collections.Counter(degree_sequence_2)
            deg_2, cnt_2 = zip(*degreeCount_2.items())
            continue
        if line.startswith('compressed_graph_without_tails:'):
            graph_str_3 = line.split('compressed_graph_without_tails: ')[1]
            graph_3 = eval(graph_str_3)
            G_3 = nx.Graph(graph_3)
            degree_sequence_3 = sorted([d for n, d in G_3.degree()], reverse=True)
            degreeCount_3 = collections.Counter(degree_sequence_3)
            deg_3, cnt_3 = zip(*degreeCount_3.items())
            continue

        if line.startswith('graph_without_bubbles:'):
            graph_str_4 = line.split('graph_without_bubbles: ')[1]
            graph_4 = eval(graph_str_4)
            G_4 = nx.Graph(graph_4)
            degree_sequence_4 = sorted([d for n, d in G_4.degree()], reverse=True)
            degreeCount_4 = collections.Counter(degree_sequence_4)
            deg_4, cnt_4 = zip(*degreeCount_4.items())
            continue
        
        

print("Сравнение степеней вершин")
print('Граф до усоврешенствований')
make_a_plot(deg_1, cnt_1)
print('Граф после сжатия')
make_a_plot(deg_2, cnt_2)
print("Граф без хвостов")
make_a_plot(deg_3, cnt_3)
print("Граф без пузырей")
make_a_plot(deg_4, cnt_4)
```

```python
import ast
import re
import matplotlib.pyplot as plt

def plot_coverage(coverage):
    values = list(coverage.values())
    plt.hist(values, bins=40)
    plt.xlabel('Покрытие')
    plt.ylabel('Частота')
    plt.show()

with open('/content/drive/MyDrive/магистратура/intermediate_calculations.txt', 'r') as f:
    for line in f:
        if line.startswith('coverage:'):
            print('Граф до усоврешенствований')
            match = re.search(r'{.*}', line)
            if match:
                dict_str = match.group()
                coverage = ast.literal_eval(dict_str)
                plot_coverage(coverage)
        elif line.startswith('compressed_coverage:'):
            print('Граф после сжатия')
            match = re.search(r'{.*}', line)
            if match:
                dict_str = match.group()
                coverage = ast.literal_eval(dict_str)
                plot_coverage(coverage)
        elif line.startswith('compressed_coverage_without_tails:'):
            print('Граф без хвостов')
            match = re.search(r'{.*}', line)
            if match:
                dict_str = match.group()
                coverage = ast.literal_eval(dict_str)
                plot_coverage(coverage)
        elif line.startswith('coverage_without_bubbles:'):
            print('Граф без пузырей')
            match = re.search(r'{.*}', line)
            if match:
                dict_str = match.group()
                coverage = ast.literal_eval(dict_str)
                plot_coverage(coverage)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import collections
import ast
import networkx as nx
from collections import defaultdict

num_vertices = []
num_edges = []

with open('/content/drive/MyDrive/магистратура/intermediate_calculations.txt', 'r') as f:
    for line in f:
        if line.startswith('graph:'):
            graph_str = line.split('graph: ')[1]
            graph = eval(graph_str)
            num_vertices.append(len(graph))
            num_edges.append(sum(len(v) for v in graph.values()))
            continue
        if line.startswith('compressed_graph:'):
            graph_str = line.split('compressed_graph: ')[1]
            graph = eval(graph_str)
            num_vertices.append(len(graph))
            num_edges.append(sum(len(v) for v in graph.values()))
            continue
        if line.startswith('compressed_graph_without_tails:'):
            graph_str = line.split('compressed_graph_without_tails: ')[1]
            graph = eval(graph_str)
            num_vertices.append(len(graph))
            num_edges.append(sum(len(v) for v in graph.values()))
            continue

        if line.startswith('graph_without_bubbles:'):
            graph_str = line.split('graph_without_bubbles: ')[1]
            graph = eval(graph_str)
            num_vertices.append(len(graph))
            num_edges.append(sum(len(v) for v in graph.values()))
            continue

x = ['Граф 1', 'Граф 2', 'Граф 3', 'Граф 4']
x_pos = [i for i, _ in enumerate(x)]

plt.bar(x_pos, num_vertices)
plt.xlabel("Графы")
plt.ylabel("Количесвто вершин")

plt.xticks(x_pos, x)
plt.show()

plt.bar(x_pos, num_edges)
plt.xlabel("Графы")
plt.ylabel("Количество ребер")
plt.xticks(x_pos, x)
plt.show()
       

```
